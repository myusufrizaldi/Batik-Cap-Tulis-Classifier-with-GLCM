```python
import imageglcm as iglcm
import rancreidml
```

```python
# First run, load_data -> train -> save

model_name = 'test_ax'
k_neighbors = 3
k_folds = 10
glcm_components = ['contrast', 'correlation', 'energy', 'homogeneity']

model = rancreidml.GlcmKnnClassifier(model_name, k_neighbors, k_folds, glcm_components)
model.load_data(is_skip=True)
model.train()
model.save()
```

```python
# Next run, load model and all set!

model_name = 'test'

model = rancreidml.GlcmKnnClassifier(model_name)
model.load()

# Predict test img
test_img = iglcm.load_img('./test/cap.jpg')
print(model.predict(test_img))

test_img = iglcm.load_img('./test/cap2.jpg')
print(model.predict(test_img))

test_img = iglcm.load_img('./test/tulis.jpg')
print(model.predict(test_img))
```

```python
# First run, load_data -> train -> save

model_name = 'test_yy'
k_neighbors = 3
k_folds = 10
glcm_components = ['contrast', 'correlation', 'energy', 'homogeneity']

model = rancreidml.GlcmKnnClassifier(model_name, k_neighbors, k_folds, glcm_components)
model.load_data()

epochs_list        = [60]
k_neighbors_list   = [k_neighbors for i in range(len(epochs_list))]
training_rate_list = [0.8 for i in range(len(epochs_list))]
iterations_list    = [10]

model.deep_train(k_neighbors_list, training_rate_list, iterations_list, epochs_list, is_accuracy_oriented=False, is_full_epochs=True, is_skip_epoch_validation=True)
model.save()
```
