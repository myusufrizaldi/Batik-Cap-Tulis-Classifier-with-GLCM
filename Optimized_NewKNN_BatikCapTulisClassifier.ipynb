```python
# First run, load_data -> train -> save

model_name = 'abcd'
k_neighbors = 3
k_folds = 10
glcm_components = ['contrast', 'correlation', 'energy', 'homogeneity']

model = rancreidml2.GlcmKnnClassifier(model_name, k_neighbors, k_folds, glcm_components)
model.load_data()

epochs_list        = [20, 30, 50, 60]
k_neighbors_list   = [k_neighbors for i in range(len(epochs_list))]
training_rate_list = [0.8 for i in range(len(epochs_list))]
iterations_list    = [10, 10, 10, 10]

model.deep_train(k_neighbors_list, training_rate_list, iterations_list, epochs_list, is_accuracy_oriented=False, is_full_epochs=True, is_skip_epoch_validation=True)
model.save()
```
